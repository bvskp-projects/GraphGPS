```python
import numpy as np
import networkx as nx

import matplotlib.pyplot as plt
```

```python
g = nx.star_graph(10)
L = nx.laplacian_matrix(g).toarray()
U, _, _ = np.linalg.svd(L)
```

```python
gp = g.copy()
gp.add_edge(np.argmin(U[-1]), np.argmax(U[-1]))
Lp = nx.laplacian_matrix(gp).toarray()
Up, _, _ = np.linalg.svd(Lp)
```

```python
plt.plot(range(11), U[-1], label="star graph")
plt.plot(range(11), Up[-1], label="one extra edge")
plt.xlabel("Node ID")
plt.ylabel("LapPE")
plt.legend()
```
